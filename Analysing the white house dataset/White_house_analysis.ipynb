# How does length of employee titles correlate to salary?


```python
import numpy as np
employee_titles = white_house["Position Title"]
employee_salary = white_house["Salary"]
white_house["Length of titles"] = employee_titles.apply(lambda x: len(x))
positionlength_salary = white_house[["Length of titles","Salary"]]
plt.scatter(positionlength_salary["Length of titles"], positionlength_salary["Salary"])
plt.xlabel("Length of title")
plt.ylabel("Salary")
plt.show()
```

A positive Correlation is seen. But with many outliers.

# What words are the most common in titles?

```python
string=''
words = string.join(white_house["Position Title"].values).split(" ")
def find_word_count(list_words):
    """
    Input: a list of strings
    Output: a Dictionary of word counts
    """
    count_words = {}
    for word in list_words:
        if word in count_words:
            count_words[word] += 1
        else:
            count_words[word] = 0
    return(count_words)

word_count_dict = find_word_count(words)
word_count_dict
```

```python
def find_max_count_word(d):
    """
    Input: A dictionary
    Output: A sorte representing the top common words in the titles.
    """
    return(sorted(d.items(), key=lambda x: x[1], reverse = True))

word_common_top = find_max_count_word(word_count_dict)

#Most common important words.
word_common_top[4:7]
```

We observe that 'Assistant', 'President' and 'Director' are the most frequent words in the position title. (All though we have 'The', 'Of' etc. but are not important, a better way is to assign TF-IDF scores but this is just a small demonstration, so avoided extra computations)
